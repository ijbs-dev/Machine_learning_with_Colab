# Imports


```python
# Importation of dependencies
import pandas as pd
from sklearn import linear_model as lm
import numpy as np
```


# Loading Data

```python
# Importation of dependencies
# file_path = 'caminho/para/o/arquivo.csv'
# df_raw = pd.read_csv(file_path)

# Load the dataset for a variable
# Dataset from = https://github.com/basilhan/datasets/tree/master
df_raw = pd.read_csv('https://raw.githubusercontent.com/basilhan/datasets/master/kc-house-data.csv')
```

```python
# Shows the first data of the dataset
df_raw.head()
```

# Data Preparation

```python
# Features
x_train = df_raw.drop(['price', 'date'], axis = 1)

#Response Variable
y_train = df_raw['price'].copy()
```

# Model Training

```python
# Consultar site (https://scikit-learn.org/stable/modules/linear_model.html#regression)

# Model description
model_lr = lm.LinearRegression()

# Model Training
model_lr.fit(x_train, y_train)
```

```python
# Prediction
pred = model_lr.predict(x_train)
```

# Performance Metrics

```python
# print 100 fist values
pred[0:100]
```

```python
df1 = df_raw.copy()
```

```python

df1['prediction'] = pred
```

```python
print('\n')
print(df1[['id', 'price', 'prediction', 'error']].head())
```

```python
# Now we create the error column to represent the MAE and the column error_abs to store the absolute value
df1['error'] = df1['price'] - df1['prediction']
df1['error_abs'] = np.abs( df1['error'] )

# Likewise, we now create the error_perc column to represent MAPE and the error_perc_abs column to store the absolute value
df1['error_perc'] = ( (df1['price'] - df1['prediction']) / df1['price'] )
df1['error_perc_abs'] = np.abs( df1['error_perc'] )

df1[['id', 'price', 'prediction', 'error', 'error_abs', 'error_perc', 'error_perc_abs']].head()
```

```python
# Mean absolute error (mae)
mae = np.mean( df1['error_abs'] )
print('MAE: {}'.format( mae ))
```

```python
# Mean absolute percentage error (mape)
mape = np.mean( df1['error_perc_abs'] )
print('MAPE: {}'.format( mape ))
```
