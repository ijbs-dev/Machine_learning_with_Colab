# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vToetCNM_P0WC6O9DXpGFEmrb3y8BcAov6H0Rx32sTtw26wbVWmHo7XjFTCruMgaZWKdymmvcBGPApG/pub?output=csv')
```

```python
# Modifica nome onde tinha 'red' troca por '0'
df['style'] = df['style'].replace('red', 0)
```

```python
# Modifica nome onde tinha 'white' troca por '1'
df['style'] = df['style'].replace('white', 1)
```

```python
df
```

```python
# Below we separate the variables between predictors and target variable
# Separando as variaveis entre preditoras e variavel alvo
y = df['style']
x = df.drop('style', axis = 1) # Está excluindo apenas coluna 'color'
```

```python
df.shape # Mostra o tamanho de 'wine_df'
```

```python
y
x
```

```python
from sklearn.model_selection import train_test_split

# Creating the train and test data sets
# Crian variaveis de treino (x_train e y_train) e teste (x_test e y_test);
# Define 30% do total pra teste (test_size = 0.3) e automaticamente 70% para treino;
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python

# Converter x_train e y_train de numpy.ndarray para DataFrame/Series
x_train_df = pd.DataFrame(x_train, columns=x.columns)
y_train_series = pd.Series(y_train)

# Encontre as colunas com valores não numéricos em x_train
non_numeric_columns = x_train_df.select_dtypes(include=['object']).columns

# Converta as colunas relevantes para o formato numérico usando LabelEncoder
label_encoder = LabelEncoder()
for column in non_numeric_columns:
    x_train_df[column] = label_encoder.fit_transform(x_train_df[column])

# Converta a coluna alvo para o formato numérico
y_train_encoded = label_encoder.fit_transform(y_train_series)

# Model creation
model = ExtraTreesClassifier()
model.fit(x_train_df, y_train_encoded)
```

```python
# Converter x_test e y_test de numpy.ndarray para DataFrame/Series
x_test_df = pd.DataFrame(x_test, columns=x.columns)
y_test_series = pd.Series(y_test)

# Encontre as colunas com valores não numéricos em x_test
non_numeric_columns_test = x_test_df.select_dtypes(include=['object']).columns

# Converta as colunas relevantes para o formato numérico usando LabelEncoder
label_encoder_test = LabelEncoder()
for column in non_numeric_columns_test:
    x_test_df[column] = label_encoder_test.fit_transform(x_test_df[column])

# Converta a coluna alvo para o formato numérico
y_test_encoded = label_encoder_test.fit_transform(y_test_series)
```

```python
result = model.score(x_test_df, y_test_encoded)
print("Accuracy:", result)
```
