### Definindo Baseline

* Uma baseline é importante para ter marcos no projeto.
* Permite uma explicação facil para todos os envolvidos.
* É algo que sempre tentaremos ganhar na medida do possivel.

```python
# Converte os dados para inteiro
data.RM = data.RM.astype(int)
```

```python
# ESTATISTICA DE RM (número médio de cômodos em uma casa)
# Estatistica descritiva da coluna numero de quartos
data.RM.describe()
```

```python
# CRIA CATEGORIAS
# Define a regra para caracterizar os dados
categorias = []
```

```python
# ALIMENTA CATEGORIAS COM REGRAS: Pequeno, Medio e Grande
# Alimenta a lista de categorias
for i in data.RM.iteritems():
  valor = (i[1])
  if valor <= 4:
    categorias.append('Pequeno')
  elif valor < 7:
    categorias.append('Medio')
  else:
    categorias.append('Grande')
```

```python
# Cria a coluna categorias
data['categorias'] = categorias
```

```python
# Imprime a contagem de categorias
data.categorias.value_counts()
```

```python
# Agrupa as categorias e calcula as médias
medias_categorias = data.groupby(by='categorias')['MEDV'].mean()
```

```python
# Visualizando a variavel medias_categorias
medias_categorias
```

```python
# Criando o dicionario com chaves medio, grande e pequeno e seus valores
dic_baseline = {'Grande': medias_categorias[0],
                'Medio': medias_categorias[1],
                'Pequeno': medias_categorias[2]}
```

```python
# Imprime dicionario
dic_baseline
```

```python
# Cria a função que retorna baseline
def retorna_baseline(num_quartos):
  if num_quartos <= 4:
    return dic_baseline.get('Pequeno')
  elif num_quartos < 7:
    return dic_baseline.get('Medio')
  else:
    return dic_baseline.get('Grande')
```

```python
# Chama a função retorna baseline
retorna_baseline(3)
```

```python
# Imprime n_quartos e o valor medio
for i in data.RM.iteritems():
  n_quartos = i[1]
  print('Número de quatos é: {} , Valor médio: {}'.format(n_quartos, retorna_baseline(n_quartos)))
```

```python
# Variavel y é nossa variavel alvo
# MEDV:Valor médio de casas ocupadas pelos proprietarios de U$ 1000
y = data['MEDV']
```

```python
# Remove as colunas colineares, variavel alvo e a coluna categorias
x = data.drop(['TOWN','TRACT','LAT','LON','RAD','TAX','MEDV','DIS','AGE','ZN','categorias'],axis=1)
```

```python
# Dados da variável x
x.head()
```

```python
# Variavel alvo
y.head()
```

```python
# Carrega o metodo train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Divide os dados entre o conjunto de treino e teste, 80% e 20% respectivamente.
# Define qualquer valor para o parametro random_state.
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=5)
```

```python
from tabulate import tabulate

# Criando uma lista com as informações dos conjuntos de dados
dados = [
    ['x_train', x_train.shape[0], x_train.shape[1] if len(x_train.shape) > 1 else 0],
    ['x_test', x_test.shape[0], x_test.shape[1] if len(x_test.shape) > 1 else 0],
    ['y_train', y_train.shape[0], y_train.shape[1] if len(y_train.shape) > 1 else 0],
    ['y_test', y_test.shape[0], y_test.shape[1] if len(y_test.shape) > 1 else 0]
]

# Imprimindo as informações em formato de tabela
print(tabulate(dados, headers=['Conjunto', 'Nº Linhas', 'Nº Colunas'], tablefmt='grid'))
```

```python
# Imprimindo as 5 primeiras linhas dos dados de teste
x_test.head()
```

```python
# Definindo uma lista de predições
predicoes = []
```

```python
# Para cada elemento do teste, alimenta a lista predições
for i in x_test.RM.iteritems():
  n_quartos = i[1]
  predicoes.append(retorna_baseline(n_quartos))
```

```python
# Imprime as 10 primeiras linhas da lista de predições
predicoes[:10]
```

### Atribuindo resultados a um dataframe

*  Manteremosum log ao longo do projeto
*  Permite a evolução dado o baseline

```python
# Cria um dataframevazio
df_results = pd.DataFrame()
```

```python
# Adiciona a coluna valor_real
df_results['valor_real'] = y_test.values
```

```python
# Cria a coluna valor_predito_baseline com as predicoes
df_results['valor_predito_baseline'] = predicoes
```

```python
# Imprimindo as 10 primeiras linhas do df_results
df_results.head(10)
```

### Apesar de nao ser Util vamos criar grafico usando DataFrame()

*  Vamos usar plotly:


```python
import plotly.graph_objects as go

# Criando Vestigios (rastros)
fig = go.Figure()

# Linha com os dados de teste
fig.add_trace(go.Scatter(x=df_results.index,
                         y=df_results.valor_real,
                         mode='lines+markers',
                         name='Valor Real'))

# Linha com os dados preditos
fig.add_trace(go.Scatter(x=df_results.index,
                         y=df_results.valor_predito_baseline,
                         mode='lines+markers',
                         name='Valor Predito Baseline'))

# Plota a figura
fig.show()
```

### RMSE (root mean squared error): é a medida que calcula "a raiz quadrática média" dos erros entre valores observados (pred = reais) e predições (ref = hipóteses).

RMSE.JPG

```python
# Calcula a metrica RMSE
from sklearn.metrics import mean_absolute_error
from math import sqrt
```

```python
# Aplicando a formula RMSE
rmse = (np.sqrt(mean_absolute_error(y_test, predicoes)))
```

```python
# Imprime a performace do modelo
print ('Performance do modelo baseline:')
print('\nRMSE é {}'.format(rmse))
```

### Machine Learning

*  Aprendizagem Supervisionada
  
  *   Regressão Linear

    *   Devido a correlação entre algumas features podemos usar a regressão linear.
    *   Modelo simples e eficiente.
    *   Solução rápida e robusta.

```python
# Carrega o pacote LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
# Cria o objeto do tipo LinearRegression
lin_model = LinearRegression()
```

```python
# Treina o algoritmo de Regressão Linear
lin_model.fit(x_train,y_train)
```

```python
# Avaliação do Modelo nos dados de teste
y_pred = lin_model.predict(x_test)
```

```python
# Calcula  a metrica RMSE
rmse = (np.sqrt(mean_absolute_error(y_test, predicoes)))
```

```python
# Imprime a performace do modelo
print ('Performance do modelo Avaliado com os dados de teste:')
print('\nRMSE é {}'.format(rmse))
```

```python
# Atribui os resultados no Dataframe df_results
df_results['valor_predito_reg_linear'] = lin_model.predict(x_test)
```
